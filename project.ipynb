### Loop through different pages

```python
import requests as r
from bs4 import BeautifulSoup

def get_urls(num):
    """Function to get all the urls of the houses for sale from different pages """
    list_all_url = [] #list containing all urls 
    count = 0   # count number of houses
    for i in range(1,num + 1):  #for loop to loop through pages
        root_url = f"https://www.immoweb.be/en/search/house/for-sale?countries=BE&page={i}&orderBy=relevance"
        req = r.get(root_url)
        content = req.content
        soup = BeautifulSoup(content, "html") 
        
        if req.status_code == 200:
        # loop to catch all the url of houses
            for tag in soup.find_all("a", attrs = {"class" : "card__title-link"}):
                list_all_url.append(tag.get("href"))
                count += 1
        # when the page number exceeds the existing page number    
        else:
            print("page not found")
            break
    print(f"Number of houses: {count}")

    return list_all_url

#get_urls()

```

```python
from concurrent.futures import ThreadPoolExecutor
import time

start_time = time.process_time()

with ThreadPoolExecutor(max_workers=20) as executor:
    result = [executor.submit(get_urls, 100)]
    
end_time =time.process_time()
print(f"Time taken to complete {round(end_time - start_time,2)} seconds")
list_urls = [item.result() for item in result]
print(list_urls)
```
