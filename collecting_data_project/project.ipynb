### Loop through different pages

```python
import requests as r
from bs4 import BeautifulSoup
list_all_url = []
for i in range(5):
    root_url = f"https://www.immoweb.be/en/search/house/for-sale?countries=BE&page={i+1}&orderBy=relevance"
    req = r.get(root_url)
    content = req.content
    soup = BeautifulSoup(content, "html")
    
    for tag in soup.find_all("a", attrs = {"class" : "card__title-link"}):
        list_all_url.append(tag.get("href"))

list_all_url

```
